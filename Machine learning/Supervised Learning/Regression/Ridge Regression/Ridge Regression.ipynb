```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor
```

```python
#cross validation(cv) - given dataset will convert multiple data set format
from sklearn.model_selection import cross_val_score
mean_square_error = cross_val_score(regressor, x, y, scoring = "neg_mean_squared_error", cv=5)
mean_square_error
```

```python
#linear regression best score
np.mean(mean_square_error)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
```

```python
#ridge regression - we will give alpha value - 1e = 10 - 10^2
parameters = {"alpha":[1e-15, 1e-10, 1e-8, 1e-5, 1e-2,1,5,10,20,30,40,50]}
ridge_regressor = GridSearchCV(ridge, parameters, scoring = "neg_mean_squared_error", cv=5, verbose=1)
```

```python
ridge_regressor.fit(x,y)
```

```python
#calculate best parameter - 
ridge_regressor.best_params_
```

```python
#ridge_regressor best score
ridge_regressor.best_score_
```

```python
#test have 0.3 remaining 0.7 go to train section
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
prediction_ridge = ridge_regressor.predict(x_test)
```

```python
#model prediction will find the r^2and adjusted r^2 - r^2 value is higher then the model is better
from sklearn.metrics import r2_score
```

```python
#find r2
r2=r2_score(y_test,prediction_ridge)
r2
```
