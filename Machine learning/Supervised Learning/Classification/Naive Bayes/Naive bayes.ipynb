```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\ShibuKumar\Downloads\Datas\emails.csv")
data.head()
```

```python
dataset = data[["Email No.","to"]]
dataset
```

```python
dataset.isnull().sum()
```

```python
y,x= dataset.to,dataset["Email No."]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
vectorizer = CountVectorizer()
count = vectorizer.fit_transform(x_train.values)
count
```

```python
classifier = MultinomialNB()
target = y_train.values
classifier.fit(count,target)
```

```python
y_predict = classifier.predict(vectorizer.transform(x_test))
```

```python
confusion_matrix(y_test,y_predict)
```

```python
accuracy_score(y_predict,y_test)    
```
