```python
#import the necessary packages
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
dataset=pd.read_csv(r"C:\Users\ShibuKumar\Downloads\Datas\loan_data.csv",sep=",",header=0)
dataset.head()
```

```python
len(dataset)
```

```python
dataset.shape
```

```python
#Seperation the target variables
x=dataset.iloc[:,2:13]
y=dataset.iloc[:,13]
```

```python
#splitting dataset into train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
#Function with perform traing to entropy
classification_entropy=DecisionTreeClassifier(criterion="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
```

```python
classification_entropy.fit(x_train,y_train)
```

```python
#Fn to make prediction
y_predict = classification_entropy.predict(x_test)
y_predict
```

```python
#checking accuracy
accuracy_score(y_test,y_predict)
```
